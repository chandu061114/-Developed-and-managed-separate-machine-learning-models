```python
features_folder_path = 'features/'
# Define folder paths
features_folder = 'features'
y_test_folder = 'y_test'
```

```python
import os
import pandas as pd

# Define the path to the features folder

# Initialize a dictionary to store the datasets
features_datasets = {}

# Iterate through each file in the features folder
for file_name in os.listdir(features_folder):
    # Check if the file is a CSV file
    if file_name.endswith('.csv'):
        # Extract the stock name from the file name
        stock_name = file_name.split('_')[0]
        # Read the CSV file and store it in the dictionary with the stock name as key
        features_datasets[stock_name] = pd.read_csv(os.path.join(features_folder, file_name))

# Print the loaded datasets
for stock_name, dataset in features_datasets.items():
    print(f'Dataset for {stock_name}:')
    print(dataset.head())  # Print the first few rows of each dataset
```

```python
import os
import pandas as pd

# Define the path to the y_test folder
y_test_folder = 'y_test'

# Initialize a dictionary to store the datasets
y_test_datasets = {}

# Iterate through each file in the y_test folder
for file_name in os.listdir(y_test_folder):
    # Check if the file is a CSV file
    if file_name.endswith('.csv'):
        # Extract the stock name from the file name
        stock_name = file_name.split('_')[0]
        # Read the CSV file and store it in the dictionary with the stock name as key
        y_test_datasets[stock_name] = pd.read_csv(os.path.join(y_test_folder, file_name))

# Print the loaded datasets
for stock_name, dataset in y_test_datasets.items():
    print(f'Dataset for {stock_name}:')
    print(dataset.head())  # Print the first few rows of each dataset
```

```python
import pandas as pd


# Merge feature datasets and y_test datasets based on 'Date'
merged_datasets = {}
for stock, feature_data in features_datasets.items():
    y_test_data = y_test_datasets[stock]
    merged_data = pd.merge(feature_data, y_test_data, on='Date', how='inner')
    merged_datasets[stock] = merged_data

# Calculate daily returns for each stock
for stock, data in merged_datasets.items():
    data['daily_returns'] = data['Close'].pct_change() * 100

# Merge daily returns with next_day_expected_returns
for stock, data in merged_datasets.items():
    data['next_day_expected_returns'] = data['next_day_expected_returns'].shift(-1)

# Calculate strategy returns based on predictions for each stock
strategy_returns = {}
for stock, data in merged_datasets.items():
    data['strategy_returns'] = data['next_day_expected_returns'] * data['daily_returns']
    strategy_returns[stock] = data['strategy_returns'].clip(lower=-1)

# You can now access the strategy returns for each stock in the `strategy_returns` dictionary.
# For example:
print(strategy_returns['HDFC'])
```
